```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("predict.csv")
df.head()
```

```python

fig, ax = plt.subplots(figsize = (8,6))
sns.set_style("ticks")
sns.regplot( ax = ax,data = df, x = "be_a", y = "be_a_RFR",ci=95 ,
            scatter_kws = {"s" :75, "alpha" : 0.5,'color' :'darkblue'}, line_kws={'linewidth':2,'color' :'limegreen'})
eq = r"$R^2 = 0.946$"
ax.text(0.4, 8.5, eq, {'color': 'black', 'fontsize': 13})
plt.tick_params(direction='in')
plt.xlabel("Standard Binding Energy Per Nucleon(MeV)",fontsize=12,fontweight ="bold")
plt.ylabel("Predicted Binding Energy Per Nucleon(RFR)",fontsize=12,fontweight ="bold")
plt.xticks(color = 'k', fontsize=10,fontweight ="bold")
plt.yticks(color = 'k', fontsize=10,fontweight ="bold")
axin = ax.inset_axes([0.55,0.1,0.4,0.4])
sns.regplot( ax = axin,data = df, x = "be_a", y = "be_a_RFR",ci=95, 
            scatter_kws = {"s" :10 , "alpha" : 0.7,'color' :'darkblue'}, line_kws={'linewidth':1.5,'color' :'limegreen'})
axin.set(xlabel=None)
axin.set(ylabel=None)
axin.set_xlim([7,9])
axin.set_ylim([7,9])
plt.savefig('rfr_reg.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(16,10))
gs = fig.add_gridspec(3,3)
gs.update(wspace=0.3, hspace=0.3)
ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])

sns.regplot( ax = ax0,data = df, x = "be", y = "BE_SEMF",
            scatter_kws = {"s" :25, "alpha" : 0.5,'color' :'darkblue'}, line_kws={'linewidth':1,'color' :'limegreen'},label = "SEMF")
ax0.set_xlabel("experimental binding energy")
ax0.set_ylabel("predicted binding energy(SEMF)")
sns.regplot( ax = ax1,data = df, x = "be", y = "be_RFR",
            scatter_kws = {"s" :25, "alpha" : 0.5,'color' :'darkblue'}, line_kws={'linewidth':1,'color' :'limegreen'},label = "RFR")
ax1.set_xlabel("experimental binding energy")
ax1.set_ylabel("predicted binding energy(RFR)")
sns.regplot( ax = ax2,data = df, x = "be", y = "be_XGB",
            scatter_kws = {"s" :25, "alpha" : 0.5,'color' :'darkblue'}, line_kws={'linewidth':1,'color' :'limegreen'},label = "XGB")
ax2.set_xlabel("experimental binding energy")
ax2.set_ylabel("predicted binding energy(XGB)")
sns.regplot( ax = ax3,data = df, x = "be", y = "be_MLP",
            scatter_kws = {"s" :25, "alpha" : 0.5,'color' :'darkblue'}, line_kws={'linewidth':1,'color' :'limegreen'},label = "MLP")
ax3.set_xlabel("experimental binding energy")
ax3.set_ylabel("predicted binding energy(MLP)")
ax0.legend();
ax1.legend();
ax2.legend();
ax3.legend();
plt.savefig('bex.png', dpi=300, bbox_inches='tight')
plt.show()
```
